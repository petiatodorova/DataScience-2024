### Let's clean more or dataset by removing rows with '1', '6', '2', because these are unclear values - 1 what? kg? lbs?

```python
values = ['1', '2', '6']
pattern = '|'.join(values)
coffee_data = coffee_data[~coffee_data.bag_weight.str.contains(pattern)]
coffee_data.bag_weight.str.slice(-3).unique()
```
