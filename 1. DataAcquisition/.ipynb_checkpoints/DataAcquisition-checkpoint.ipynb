### Scraping

#### Sample 1

```python
countries_page = requests.get("https://www.scrapethissite.com/pages/simple/")
```

```python
type(countries_page)
```

```python
parser = BeautifulSoup(countries_page.content, "html.parser")
```

```python
rows = parser.find(id = "countries").find_all("div", class_="row")
```

```python
rows[10].find_all("h3", class_ ="country-name")
```

#### Sample 2

```python
url = 'https://www.scrapethissite.com/pages/forms/'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup.prettify)
```

#### Finds the first div tag in the html

```python
soup.find('div')
```

#### Finds all div tags in the html

```python
soup.find_all('div')
```

#### Finds specific div tag --> in the example with class="col-md-12"

```python
soup.find_all('div', class_ = 'col-md-12')
```

```python
soup.find_all('p', class_ = 'lead')
```

#### Find text of tag 'p' with class 'lead'

```python
soup.find('p', class_ = 'lead').text.strip()
```

```python
soup.find_all('td')
```

```python
soup.find('th').text.strip()
```
